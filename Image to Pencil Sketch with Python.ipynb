Task3-Image to Pencil Sketch with Python

Importing libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Loading Real Image

```python
image = cv2.imread("E:/smriti.jpg")
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

Converting Image to GrayScale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray_image,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

Inverting The Image

```python
inverted_image = 255 - gray_image
plt.imshow(cv2.cvtColor(inverted_image,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

Blurr The Image

```python
blurr_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

Inverting the Blurred Image

```python
inverted_blurred = 255 - blurr_image
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
plt.imshow(cv2.cvtColor(pencil_sketch,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show() 
```
